# 1. Реализуем загрузку данных и предобработку для построения таблицы Excel с количеством пар навыков

- #### 1.1 Загружаем данные из файла

```python
# Создаем массив для хранения данных
jobs_skills = []

# Открываем файл для чтение
with open('dataModuleB.txt', 'r') as f:
    # Считем каждую строчку файлы
    for value in f:
        # Разбиваем строку на массив
        items = value.strip()[1:-1].split(', ')
        # Сохраняем в массив
        jobs_skills.append(items)
```

- #### 1.2 Считаем количество пар

```python
# Из библиотеки импортируем модуль для подсчета пар
from collections import defaultdict

# Создаем словарь для подсчета пар
pair_counts = defaultdict(int)

# Циклом считаем количество вхождения навыков вместе 
for job in jobs_skills:
    for i in range(len(job)):
        for j in range(i+1, len(job)):
            # Сортируем для учета перестановок
            pair = tuple(sorted([job[i], job[j]]))
            # Увеличиваем значение пары
            pair_counts[pair] += 1


# Создаем счетчик
count = 0
# Циклом проходимся по всем парам и выводим пример
for pair, count_p in pair_counts.items():
    print(f'Навыки {pair[0]} и {pair[1]} встречаются вместе -> {count_p} раз')
    if count == 10:
        break 
    count += 1
```

# 2. Работем с сохранением данных в Excel

```python
# Импортиурем библиотеку для работы с данными
import pandas as pd
# Импортируем библиотеку для работы со строками
import re

# Создаем массив для хранения ключей
keys = []
# Создаем массив для хранения значений
values = []

# Циклом проходимся по всем данным 
for key, value in pair_counts.items():
    # Заполняем массивы
    keys.append(key)
    values.append(value)

# Подготавливаем данные для DataFrame
df_data = {'Столбец 1': [re.sub(r'[\'"]', r'', key[0]) for key in keys], 
           'Столбец 2': [re.sub(r'[\'"]', r'', key[1]) for key in keys], 
           'Столбец 3': values}

# Сохраняем данные в DataFrame
df = pd.DataFrame(df_data)

# Созраняем данные в Excel
df.to_excel('data.xlsx', index=False)
```
