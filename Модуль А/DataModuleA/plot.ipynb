# 1. Реализация графа плотности распределения

```python
# Импортиурем библиотеку для работы с данными
import pandas as pd
# Импортиурем библиотеку для работы с графиками 
import matplotlib.pyplot as plt
```

```python
# Загружаем данные из CSV в DataFrame
df = pd.read_csv(r'data/cleardata/csvcleardata.csv', encoding='utf-8')

# Создаем граф плотности распсределения 
df.plot()

# Задаем графу название
plt.title('Плотность распределения')
# Задаем название оси X
plt.xlabel('Значение')
# Задаем название оси Y
plt.ylabel('Плотность')

# Показываем граф
plt.show()
```
