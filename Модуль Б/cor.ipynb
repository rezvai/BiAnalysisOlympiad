# 1. Реализация корреляцинного анализа

```python
# Импортиурем библиотеку для рабоыты с данными
import pandas as pd
# Импортируем из библиотеки модуль для реализации корреляцинного анализа Спирмена
from scipy.stats import spearmanr
# Импортируем библиотеку для работы с графиками
import matplotlib.pyplot as plt
```

```python
# Загружаем данные из CSV в DataFrame
df = pd.read_csv(r'data/csvcleardata.csv', encoding='utf-8')

# Выводим пример DataFrame
df
```

```python
# Выводим корреляционный анализ двух переменных: заработная плата и опыт работы
print(f"Корреляционный анализ двух переменных: заработная плата и опыт работы - {spearmanr(df['salary'], df['work_experience'])}", end='\n\n')
# Выводим корреляционный анализ двух переменных: заработная плата и название региона
print(f"Корреляционный анализ двух переменных: заработная плата и название региона - {spearmanr(df['salary'], df['region_name'])}", end='\n\n')
# Выводим корреляционный анализ двух переменных: заработная плата и график работы
print(f"Корреляционный анализ двух переменных: заработная плата и график работы - {spearmanr(df['salary'], df['work_schedule'])}", end='\n\n')
# Выводим корреляционный анализ двух переменных: график работы и опыт работы
print(f"Корреляционный анализ двух переменных: график работы и опыт работы - {spearmanr(df['work_schedule'], df['work_experience'])}", end='\n\n')
# Выводим корреляционный анализ двух переменных: название региона и график работы
print(f"Корреляционный анализ двух переменных: название региона и график работы - {spearmanr(df['region_name'], df['work_schedule'])}", end='\n\n')
# Выводим корреляционный анализ двух переменных: название региона и опыт работы
print(f"Корреляционный анализ двух переменных: название региона и опыт работы - {spearmanr(df['region_name'], df['work_experience'])}")
```

### Пояснение к выводу 
Параметр statistic - показывает линейную зависимость между двумя переменными в диапозоне от -1 до 1, чем ближе значение к 1, тем выше линейная зависимость между двумя переменными.
